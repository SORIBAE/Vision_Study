```python
import math
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-math.pi, math.pi, 1000) 
```

```python
y = torch.sin(x) 
```

```python
a = torch.randn(())
b = torch.randn(())
c = torch.randn(())
d = torch.randn(())
```

```python
y_random = a * x**3 + b * x**2 + c * x + d # 사인 함수를 근사할 3차 다항식 정의
```

```python
## 그래프 출력하기
# 실제 y값으로 사인곡선 만들기
plt.subplot(2, 1, 1)
plt.title("y true")
plt.plot(x, y)

# 임의의 가중치로 만든 예측용 y값으로 사인곡선 만들기

plt.subplot(2, 1, 2)
plt.title("y radnom")
plt.plot(x, y_random)

# 그래프 출력하기
plt.show()
```
